```python
# underfit 1
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from matplotlib import pyplot
from numpy import array
```

```python
def get_train():
    seq = [[0.0,0.1],[0.1,0.2],[0.2,0.3],[0.3,0.4],[0.4,0.5]]
    seq = array(seq)
    X,y = seq[:,0], seq[:,1]
    X = X.reshape((len(X),1,1))
    return X,y
```

```python
def get_val():
    seq = [[0.5,0.6], [0.6,0.7],[0.7,0.8],[0.8,0.9],[0.9,1.0]]
    seq = array(seq)
    X,y = seq[:,0], seq[:,1]
    X = X.reshape((len(X),1,1))
    return X, y
```

```python
model = Sequential()
model.add(LSTM(10,input_shape=(1,1)))
model.add(Dense(1,activation='linear'))
model.compile(loss='mse',optimizer='adam')
```

```python
X,y = get_train()
valX, valY = get_val()
history = model.fit(X,y,epochs=100, validation_data=(valX,valY),shuffle=False)
```

```python
pyplot.plot(history.history['loss'])
pyplot.plot(history.history['val_loss'])
pyplot.title('model train vs validation loss')
pyplot.ylabel('loss')
pyplot.xlabel('epoch')
pyplot.legend(['train','validation'],loc='upper right')
pyplot.show()
```

```python
# 5 repeats
```

```python
from pandas import DataFrame
```

```python
train = DataFrame()
val = DataFrame()
for i in range(5):
    model = Sequential()
    model.add(LSTM(10,input_shape=(1,1)))
    model.add(Dense(1,activation='linear'))
    model.compile(loss='mse',optimizer='adam')
    X,y = get_train()
    valX,valY = get_val()
    history = model.fit(X,y,epochs=300, validation_data=(valX,valY),shuffle=False)
    train[str(i)] = history.history['loss']
    val[str(i)] = history.history['val_loss']
    
pyplot.plot(train, color='blue',label='train')
pyplot.plot(val, color='orange',label='validation')
pyplot.title('model train vs validation loss')
pyplot.ylabel('loss')
pyplot.xlabel('epoch')
pyplot.show()
```

```python
# grid search memory cells
```

```python
def fit_model(n_cells):
    model = Sequential()
    model.add(LSTM(n_cells, input_shape=(1,1)))
    model.add(Dense(1,activation='linear'))
    model.compile(loss='mse',optimizer='adam')
    X,y = get_train()
    history = model.fit(X,y,epochs=500,shuffle=False,verbose=0)
    valX, valY = get_val()
    loss = model.evaluate(valX,valY,verbose=0)
    return loss
```

```python
params = [1,5,10]
n_repeats=5
scores = DataFrame()
for value in params:
    loss_values = list()
    for i in range(n_repeats):
        loss = fit_model(value)
        loss_values.append(loss)
        print('>%d%d param=%f, loss=%f' %(i+1,n_repeats,value,loss))
    scores[str(value)] = loss_values
print(scores.describe())
scores.boxplot()
pyplot.show()
```

```python
# grid search batch size
```

```python
def fit_model(n_batch):
    model = Sequential()
    model.add(LSTM(10,input_shape=(1,1)))
    model.add(Dense(1,activation='linear'))
    model.compile(loss='mse',optimizer='adam')
    X,y = get_train()
    history = model.fit(X,y,epochs=500,shuffle=False,verbose=0,batch_size=n_batch)
    valX, valY = get_val()
    loss = model.evaluate(valX, valY, verbose=0)
    return loss
```

```python
params = [1,2,3]
n_repeats = 5
scores = DataFrame()
```

```python
for value in params:
    loss_values = list()
    for i in range(n_repeats):
        loss = fit_model(value)
        loss_values.append(loss)
        print('>%d/%d param=%f, loss=%f'%(i+1,n_repeats,value,loss))
    scores[str(value)] = loss_values
print(scores.describe())
scores.boxplot()
pyplot.show()
```
